## **Filtering & Sorting**

Filter orders where Category is  "Gadgets" and TotalPrice >150

```python
import pandas as pd
sales = pd.read_csv('sales_data_200.csv')
sales.dropna (inplace= True)
sales.columns = sales.columns.str.strip()
filtered_orders = sales[(sales['Category'] == 'Gadgets') & (sales['TotalPrice'] > 150)]
filtered_orders
```

**Sort the data by TotalPrice in descending order**

```python
sales_sorted = sales.sort_values (by= 'TotalPrice', ascending= False)
sales_sorted
```

**Find the top 10 orders by TotalPrice**

```python
# If you just wanted to see the boolean mask for the top 10 orders
sales_sorted.head(10)['TotalPrice'] >=150
```

```python
# To get the top 10 orders and then filter those where TotalPrice >= 150
top_10_orders = sales_sorted.head(10)
filtered_top_10_orders = top_10_orders[top_10_orders['TotalPrice'] >= 150]
print(filtered_top_10_orders)
```

```python
# Find the last 5 orders in terms of Date
# Sorts the dataset by the date Column
# `ascending=True` means oldest to the newest
# Saves the sorted data in a new table `date_asc`

date_asc = sales.sort_values(by='Date', ascending=True)
date_asc.tail(5)
```

## **Task 8: Aggregation & Analysis**

`sales.groupby('Region)` - means  groups data based on the Region column

```python
1. # Count total sales per Region

total_price_per_region = sales.groupby('Region')['TotalPrice'].sum()
print(total_price_per_region)
```

```python
2. # Find the average TotalPrice per product

total_price_by_product = sales.groupby ('Product')['TotalPrice'].sum()
print(total_price_by_product)
```

```python
3. # Alternatively you can also use mean
# Notice the total is in decimal places
# Average is often not a whole number thats why it shows decimals
total_price_by_product = sales.groupby ('Product')['TotalPrice'].mean()
print(total_price_by_product)
```

```python
# To remove the decimal places

total_price_by_product = sales.groupby ('Product')['TotalPrice'].mean()
print(total_price_by_product.astype (int))
```

```python
4. # Count total number of Orders each Customer has placed

customer_order_count = sales.groupby('CustomerName')['OrderID'].count()
print(customer_order_count)
```
