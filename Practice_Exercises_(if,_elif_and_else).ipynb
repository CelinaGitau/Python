1. **Categorise a person based on age using user input**

```python
#Categorise a person based on age

#User input your age
age = int(input("Enter your age:"))


# Check multiple conditions using if, elif, and else
if age < 15:
 print ("child")

elif age <= 19:
 print ("Teenager")

else:
 print (" You are an Adult")
```

3. **Give advice based on temperature**

4. **Simple Calculator Code**

     The code asks the user to enter two numbers

     Then asks which operation to use: **`+`** , **`-`** , **`*`**(multiply) or **`/`**(division)

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
operation = input("Enter operation (+, -, *, /): ")

if operation == "+":
    print(num1 + num2)
elif operation == "-":
    print(num1 - num2)
elif operation == "*":
    print(num1 * num2)
elif operation == "/":
    print(num1 / num2)
else:
    print("Invalid operation")
```

  2. **Ticket Price Based on Age and Day**

     Includes a nested **`if`** ( an **`if`**  inside another **`if`**)

    - If you are under 12 - Ticket costs £5

    - If you are between 12 and 59:
      - On weekend - £12
      - On weekday - £10

    - If you are 60 or older- Ticket costs 7

```python
# User enters their age, then asks if it's a weekday or weekend

age = int(input("Enter your age: "))
day = input("Enter day (weekday/weekend): ")

if age < 12:
    print("Ticket price: £4")
elif age >= 12 and age < 60:
    if day == "weekend":
        print("Ticket price: £12")
    else:
        print("Ticket price: £10")
else:
    print("Ticket price: £6")
```
