# **Python-if-elif-else**

 Designed for **Python Beginners** to learn how to use conditional statements  ('if', 'elif', 'else')


---

**if**:  checks the main condition

**elif**:  checks additional conditions if **if** is false

**else**: executes when all conditions are false



```python
a = 8
if a > 6:
  print ('a is greater than 6')
```

Using else for  fallback
**if ** condition is not true . You can use **else**



```python
a = 3
if a > 4:
    print ('a is greater than 4') #if condition 3 > 4 is false
else:
  print ('a is not greater than 4')
```

Checking Multiple Conditions Using (**elif**)

elif - checks additional conditions

```python
a = 10
if a > 15: # 10 > 15 is false
 print('a is greater than 15')
elif  a == 15: # checks if condition is false
 print ('a is equal to 15')
else:  # all previous conditions were false
  print ('a is less than 15')
```

## **Fizzbuzz Task**

 Integers from 1 to 100

 if number is divisible by 3, print "fizz"

  If number is divisible by 5, print "buzz"

   If number is both divisible by 3 and by 5, print "fizzbuzz"

   Otherwise, print just the number

```python
for number in range(1, 100):
  if number % 3 == 0 and number % 5 == 0:
    print("fizzbuzz")
  elif number % 3 == 0:
    print("fizz")
  elif number % 5 == 0:
    print("buzz")
  else:
    print(number)
```
