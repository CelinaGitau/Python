```python
import pandas as pd
income_data = pd.read_csv ('income.datascatterplot.csv')
```

### **Basic plot**

`x axis` ( Horizontal) - Income

`y axis` (Vertical) - Happiness

```python
income_data.plot (x = 'income', y= 'happiness')
```

### **Now we will create a scatter plot**
- To to show the relationship between income and happiness.
- Specifying the color of the points
- Code only shows dots and no trend line or regression line

```python
income_data.plot (x ='income', y = 'happiness', kind = 'scatter', color = 'orange')
```

### **Import Seaborn Library**

- Used for statistical graphs
- Created a scatter plot & trend line/regressionline
- To show the realtionship between Happiness and Income

```python
import seaborn as sns
sns.regplot (data =income_data, x='income', y= 'happiness', color ='green')
```



### **Scatter Plot with a Title**


```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot (data =income_data, x='income', y= 'happiness', color ='green')
plt.title ('Relationship between Income and Happiness')
plt.xlabel ('Income')
plt.ylabel ('Happiness')

plt.show ()
```

### **Histogram**

It shows how income values are distributed

**`bins=15`**  - splits data into 15 ranges


```python

import matplotlib.pyplot as plt
income_data['income'].plot(kind='hist', bins=15, color='purple')
plt.title ('Relationship between Income and Happiness')
plt.xlabel ('Income')
plt.ylabel ('Happiness')
plt.show ()
```

### **Box plot**

Provides a different type of visualisation, but a scatter or regression plot best shows the relationship




```python
income_data[['income', 'happiness']].plot(kind='box')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=income_data[['income', 'happiness']], palette='Set2')

plt.title('Income vs Happiness Box Plot')
plt.show()
```

###  **Correlation Heatmap**

🔴 =  Strong relationship

🔵 =  Weaker relationship

The closer to 1, stronger the relationship

```python
import seaborn as sns

sns.heatmap(income_data.corr(), annot=True, cmap='coolwarm')
```
