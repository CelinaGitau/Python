# Learning Python **LOOPS:**


Create a variable called **x**,  starting value of 1

Ask the Question: Is **x**, less than 10?

If the answer is **yes**, the code inside loop runs
If the answer is **no**,the loop stops

```python

x = 1  # created variable, starting value 1
while x <10: # runs loop as long as x is less than 10
  print(x)
  x += 1    #adds 1 to x each time
```

# **Method 1:**

**Write a loop and print all even numbers from 0 to 30**




---



 Create a variable  called  **number** (your choice), starting values of 0

 Is the number less than or equal to 30?

 If **yes** - keep running the Loop

 If **no** - Stop the loop


```python
number = 0
while number <= 30:
  print (number)
  number += 2
```


# **Method 2: using  continue condition in Python:**

**Write a loop and print all even numbers from 0 to 30 **


---



 Is the number less than or equal to 30?

 If **yes** - keep running/ continue the Loop

 If **no** - Stop the loop


```python
number = 0
while number <= 30:
  number += 2
  if number==32:
    continue
  print(number)
```

# **Method 3: Using the break condition in Python**

**Write a loop and print all even numbers from 0 to 30 **


---


 If **yes** - keep running the Loop

 If **no** - break the loop

```python
number = 0
while number <= 30:
  number+=2
  if number == 32:
    break
  print(number)
```


# **Code Method 4 using % :**
